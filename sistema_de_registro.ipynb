```python
def main():
    nome_usuario = input("Bom dia! Qual é o seu nome? ")

    numero_computadores = int(input(
        f"Ok {nome_usuario}, quantos computadores você precisa alugar? "
    ))

    horas_de_uso = int(input("Por quantas horas irá usar? "))

    TAXA_POR_HORA = 10
    valor_total = numero_computadores * horas_de_uso * TAXA_POR_HORA

    forma_pagamento = input(
        f"Valor total do serviço: R${valor_total:.2f}. "
        "Qual será a forma de pagamento? "
    )

    print(f"Pagamento via {forma_pagamento} registrado. Muito obrigado! Volte sempre.")

if __name__ == "__main__":
    main()
```
