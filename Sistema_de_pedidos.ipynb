```python
# -*- coding: utf-8 -*-

CARDAPIO = {
    "1": "Cachorrão",
    "2": "Cachorrinho",
    "3": "Tortinha",
    "4": "Combo Cachorrão",
    "5": "Combo Cachorrinho",
    "6": "Combo Tortinha",
}

FORMAS_PAGAMENTO = {
    "1": "Pix",
    "2": "Cartão",
    "3": "Dinheiro",
}


def escolher_item():
    print("\nEscolha o número do pedido:")
    for numero, nome in CARDAPIO.items():
        print(f"{numero}) {nome}", end="  ")
    print()

    while True:
        item = input("Qual é o número do seu pedido? ").strip()
        if item in CARDAPIO:
            return CARDAPIO[item]
        print("Opção inválida. Escolha um dos números indicados.")


def perguntar_continuar():
    while True:
        resposta = input("Algo mais? (Sim/Não) ").strip().lower()
        if resposta in ("sim", "s"):
            return True
        elif resposta in ("não", "nao", "n"):
            return False
        print("Resposta inválida. Digite Sim ou Não.")


def escolher_pagamento():
    print("\nForma de pagamento:")
    for numero, nome in FORMAS_PAGAMENTO.items():
        print(f"{numero}) {nome}", end="  ")
    print()

    while True:
        opcao = input("Escolha uma opção: ").strip()
        if opcao in FORMAS_PAGAMENTO:
            return FORMAS_PAGAMENTO[opcao]
        print("Opção inválida.")


def fazer_pedido():
    print("Boa noite! Vamos montar seu pedido.")
    itens = []

    while True:
        itens.append(escolher_item())
        if not perguntar_continuar():
            break

    pagamento = escolher_pagamento()

    pedido = {
        "Itens": itens,
        "Pagamento": pagamento,
    }

    return pedido


def main():
    pedido = fazer_pedido()
    print("\n--- Resumo do pedido ---")
    print(pedido)


if __name__ == "__main__":
    main()
```
