```python
import random


def jogar_cara_coroa():
    print("\n--- Cara ou Coroa ---")
    opcoes = ["cara", "coroa"]
    jogada = input("Cara ou Coroa? ").strip().lower()

    if jogada not in opcoes:
        print("Opção inválida. Escolha 'Cara' ou 'Coroa'.")
        return

    resultado = random.choice(opcoes)
    print(f"Resultado: {resultado.capitalize()}")

    if jogada == resultado:
        print("Você ganhou!")
    else:
        print("Você perdeu!")


def jogar_pedra_papel_tesoura():
    print("\n--- Pedra, Papel ou Tesoura ---")
    opcoes = ["pedra", "papel", "tesoura"]
    jogada = input("Pedra, Papel ou Tesoura? ").strip().lower()

    if jogada not in opcoes:
        print("Opção inválida.")
        return

    resultado = random.choice(opcoes)
    print(f"Resultado: {resultado.capitalize()}")

    if jogada == resultado:
        print("Empate!")
        return

    # regras: chave vence valor
    vence_de = {
        "pedra": "tesoura",
        "papel": "pedra",
        "tesoura": "papel",
    }

    if vence_de[jogada] == resultado:
        print("Você ganhou!")
    else:
        print("Você perdeu!")


def jogar_par_impar():
    print("\n--- Par ou Ímpar ---")
    escolha = input("Você quer ser Par ou Ímpar? ").strip().lower()

    if escolha not in ("par", "impar", "ímpar"):
        print("Opção inválida. Escolha 'Par' ou 'Ímpar'.")
        return

    try:
        jogada = int(input("Jogue seu número (0 a 10): "))
    except ValueError:
        print("Digite um número válido.")
        return

    resultado = random.randint(0, 10)
    soma = jogada + resultado
    print(f"Você jogou {jogada}, o computador jogou {resultado}. Soma: {soma}")

    soma_e_par = soma % 2 == 0
    usuario_escolheu_par = escolha == "par"

    if soma_e_par == usuario_escolheu_par:
        print("Você ganhou!")
    else:
        print("Você perdeu!")


def main():
    jogar_cara_coroa()
    jogar_pedra_papel_tesoura()
    jogar_par_impar()


if __name__ == "__main__":
    main()
```
